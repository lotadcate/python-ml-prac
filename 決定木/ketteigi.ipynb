```python
import pandas as pd
import numpy as np
```

```python
df_past = pd.read_csv("4-5_user_data.csv")
df_future = pd.read_csv("4-5_user_data_future.csv")
```

```python
df_past.columns
```

```python
X = df_past[['sex', 'student', 'stay time']]
y = df_past['registration']
```

```python
from sklearn import tree
from dtreeviz.trees import *
import graphviz
```

```python
# 分類問題
dtree = tree.DecisionTreeClassifier(max_depth=2)
dtree.fit(X, y)

viz = dtreeviz(dtree, X, y, 
              target_name = 'registration',
              feature_names = ['sex', 'student', 'stay time'],
              class_names = ["not register", "register"], # データの0, 1に合わせる
              X = [1,1,50]) # データを代入して、どこに属するか予想
viz
```

```python
df_past.head()
```

```python
# 何十人も一気に予測
dtree.predict(df_future)
```

```python
dtree = tree.DecisionTreeClassifier(max_depth=4)
dtree.fit(X, y)

viz = dtreeviz(dtree, X, y, 
              target_name = 'registration',
              feature_names = ['sex', 'student', 'stay time'],
              class_names = ["not register", "register"], # データの0, 1に合わせる
              X = [1,1,50]) # データを代入して、どこに属するか予想
viz
```

```python
# 回帰問題
df_past = pd.read_csv("4-6_sales.csv")
df_future = pd.read_csv("4-6_sales_future.csv")

X = df_past[['temperature', 'price', 'rainy']]
y = df_past['sales']
            
dtree = tree.DecisionTreeRegressor(max_depth=2)
dtree.fit(X, y)

viz = dtreeviz(dtree, X, y, 
              target_name = 'sales',
              feature_names = ['temperature', 'price', 'rainy'],
              X = [20, 150 ,0]
              )
viz
```

```python
dtree.predict(df_future)
```
